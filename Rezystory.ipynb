```python
from pulp import LpMaximize, LpProblem, LpVariable
```

```python
resistances = [8, 6, 4, 10, 8]
max_currents = [2, 3, 4, 2, 2]

# Problem optymalizacyjny
prob = LpProblem("Maximize_Current", LpMaximize)

currents = [LpVariable(f"I{i+1}", 0, max_currents[i]) for i in range(5)]


prob += currents[2]
prob += currents[0] + currents[1] == currents[2]          # i1 + i2 == i3
prob += currents[2] == currents[3] + currents[4]          # i3 == i4 + i5
prob += resistances[0] * currents[0] == resistances[1] * currents[1]  # r1 * i1 == r2 * i2
prob += resistances[3] * currents[3] == resistances[4] * currents[4]  # r4 * i4 == r5 * i5


prob.solve()

results = {
    "I1": round(currents[0].varValue, 2),
    "I2": round(currents[1].varValue, 2),
    "I3": round(currents[2].varValue, 2),
    "I4": round(currents[3].varValue, 2),
    "I5": round(currents[4].varValue, 2)
}

print("Wyniki optymalizacji:")
for variable, value in results.items():
    print(f"{variable} = {value} A")

max_current = round(I3.varValue, 2)
print(f"Maksymalny prąd w układzie to: {max_current} A")
```
