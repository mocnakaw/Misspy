```python
import pulp
```

```python

resistances = [8, 6, 4, 10, 8]
currents_nominal = [4, 2, 2, 2, 4]
voltages = [6, 10, 4, 7, 3]
delta_I = 1

variables = ["I1", "I2", "I3", "I4", "I5"]
currents = {}

# Zmienne prądy w układzie
for i in range(5):
    currents[variables[i]] = pulp.LpVariable(variables[i], lowBound=currents_nominal[i] - delta_I,
                                              upBound=currents_nominal[i] + delta_I)

# Problem optymalizacji
prob = pulp.LpProblem("Minimize_Power_Dissipation", pulp.LpMinimize)

# Minimalizacja mocy rozpraszanej
power_dissipation = pulp.lpSum([voltages[i] * currents[variables[i]] for i in range(5)])
prob += power_dissipation, "Moc rozpraszana"


prob.solve()


print(f"Minimalna moc rozproszona: {pulp.value(prob.objective)} mW")
for var in prob.variables():
    print(f"{var.name} = {var.varValue} mA")
```
